# Conjectura de Collatz 

A Conjectura de Collatz consiste no seguinte procedimento: 

\begin{align}
        \text{f}(n) = \left\{
        \begin{array}{cl}
        \frac{n}{2} & \text{ se } n \text{ for par } \\
        3n + 1 & \text{ se } n \text{ for ímpar }.
        \end{array}
        \right.
    \end{align}

#Objetivos 

1.   Escrever um código para calcular a sequência de Collatz para um valor inicial arbitrário. 
2.   Incluir uma rotina para plotar um gráfico a cada iteração. 
3. Verificar o comportamento com diferentes valores iniciais. Qual o maior número de iterações que se consegue obter considerando apenas números iniciais menores que 1000? 

# Resolução da função principal



```python
def col(n): 
  vetor = [] # vetor para que os valores sejam guardados 
  while True: # crianção de um loop infinito 
    vetor.append(n) # guardando todos os valores de saída 
    if n%2 ==0 and n>1: # fazendo a primeira condição de se n>1 e n for divisível por 2, divide-se esse mesmo n por 2 
      n = n/2 
    elif n%2!=0 and n >1:  #fazendo a segunda condição de se n>1 e n for ímpar, multiplica-se esse n por 3 e soma 1
      n = 3*n + 1  
    if n == 1: # quando n for igual a 1, o loop se fecha e adiciona o 1, que é a última saída
      vetor.append(n)
      return vetor
      break

n1 = int(input('n1:')) # entrada de um número n que seja natural. 
print("sequencia 1:",col(n1))

n2 = int(input('n2:')) 
print("sequencia 2:",col(n2)) 

n3 = int(input('n3:')) 
print("sequencia 3:",col(n3))

n4 = int(input('n4:')) 
print("sequencia 4:",col(n4))
```

# Gráficos

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(col(n1), linewidth=3.0)
plt.title('Sequência 1', size = 20)
plt.grid(color='gray')

plt.subplot(2, 2, 2)
plt.plot(col(n2), color ='orange', linewidth=3.0)
plt.title('Sequência 2', size = 20)
plt.grid(color='gray')

plt.subplot(2, 2, 3)
plt.plot(col(n3), color = 'green', linewidth=3.0)
plt.title('Sequência 3', size = 20)
plt.grid(color = 'gray')

plt.subplot(2, 2, 4)
plt.plot(col(n4), color = 'red', linewidth=3.0)
plt.title('Sequência 4', size = 20)
plt.grid(color = 'gray')

plt.tight_layout()
plt.show()
```

# Maior número de iterações com valores iniciais menores que 1000

```python
# o maior número de iterações menores que 1000 
vet_novo = [] # cria-se um novo vetor 
for i in range(1,1000): # percorre um for de 1 até 999 
  a = len(col(i)) #conta, através do tamanho do vetor, o número de iterações para cada i e atribui a variável a 
  vet_novo.append(a) # adiciona o "a" ao vetor novo 
p = max(vet_novo) # atribui a variavel p o valor maximo do vetor novo 
print('O maior número de iterações é:',p)
```

# Outra forma de visualização - Gráfico Granizo 

<h5> O gráfico granizo representa o número de iterações com relação ao número de entrada n. 
O mesmo é chamado de gráfico granizo pois o granizo antes de chegar a um ponto, vai de um lugar ao outro.</h5>



```python
#gráfico granizo 
plt.figure(figsize=(15, 10))
vet_novo_1 = []
for i in range(len(vet_novo)):
  vet_novo_1.append(i)
plt.scatter(vet_novo_1, vet_novo, marker = '*', linewidths = 2)
plt.title('Gráfico Granizo', size = 25)
plt.xlabel('n', size = 20)
plt.ylabel('Iterações', size = 20 )
plt.grid(color='purple')
```
